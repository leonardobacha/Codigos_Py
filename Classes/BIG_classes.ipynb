```python
#PARA BIG
#função que realiza gráfico
#função que define anotações metabólicas

fec_lista = ['Bacillus', 'Clostridium', 'Enterococcus', 'Klebsiella', 'Vibrio',
                'Escherichia','Cloacibacterium', 'Bacteroides', 'Enterobacter', 'Enterococcus',
                'Faecalibacterium', 'Citrobacter', 'Akkermansia', 'Parabacteroides', 'Anaerostipes',
                'Anaerotruncus', 'Coprococcus', 'Dorea', 'Bifidobacterium', 
                'Eubacterium', 'Methanobrevibacter', 'Clostridioides', 'Enterobacter', 'Enterocloster','Faecalicatena',
                'Shigella','Methanosphaera', 'Providencia', 'Pleisiomonas', 'Arachnia', 'Eggerthella',
                'Salmonella', 'Staphylococcus', 'Streptococcus', 'Campylobacter', 'Listeria', 'Helicobacter', 
                'Yersinia', 'Pseudomonas', 'Aeromonas', 'Eubacterium', 'Leptospira', 'Serratia', 'Arcobacter']


class Coleta:
    
    def __init__(self, cod, gen, arg, met, data):
        self.cod = cod
        self.gen = gen
        self.arg = arg
        self.data = data      
        self.met = met
    
    def fec_list(self,list_gen):
        j = list_gen[0]
        for i in range(1, len(list_gen)):
            j = pd.merge(j, list_gen[i], on=['gen'], how = 'outer')
        x = []
        for index, col in j.iterrows():
            if col['gen'] in fec_lista:
                x.append(col['gen'])
        j.index = j['gen']
        fec = j.loc[x,:]
        fec['mean'] = round(fec.mean(axis=1),2)
        fec['std'] = round(fec.std(axis=1))
        fec = fec.drop(columns ='gen')
        return fec
    
#Definir uma função que agrupe apenas os generos, e uma que peça ao usuario para inserir a lista de generos desejada 
    
    def arg_list(self,list_arg):
        j = list_arg[0]
        for i in range(1, len(list_arg)):
            j = pd.merge(j, list_arg[i], on=['gen'], how = 'outer')
        j.index = j['gen']
        arg = j
        arg['mean'] = round(arg.mean(axis=1),2)
        arg['std'] = round(arg.std(axis=1))
        arg = arg.drop(columns ='gen')
        return arg

    def met_list(self,list_met):
        j = list_met[0]
        for i in range(1, len(list_met)):
            j = pd.DataFrame.join(j, list_met[i])
        j.index = ['Amino Acids and Derivatives','Carbohydrates',"Cofators,Vitamins,Prosthetic Groups",
                  "FattyAcids,Lipids,and Isoprenoids",
                  "Iron acquisition and metabolism","Metabolism damage and its repair or mitigation",
                  "Nitrogen Metabolism","Nucleosides and Nucleotides",
                  "Phosphate Metabolism","Secondary Metabolism","Sulfur Metabolism"]

        met = j
        met['mean'] = round(met.mean(axis=1),2)
        met['std'] = round(met.std(axis=1))
        return met
    
class Rio:
    
    def __init__(self, nome_rio, coleta_1,coleta_2,coleta_3):
        self.nome_rio =nome_rio
        self.coletas = [coleta_1.cod,coleta_2.cod,coleta_3.cod]
        self.arg = [coleta_1.arg,coleta_2.arg,coleta_3.arg]
        self.gen = [coleta_1.gen,coleta_2.gen,coleta_3.gen]
        self.met = [coleta_1.met,coleta_2.met,coleta_3.met]
    # def agrupar_ciano(self):
       
    def agrupar_fec(self):
        return Coleta.fec_list(self,self.gen) 
    
    def agrupar_arg(self):
        return Coleta.arg_list(self,self.arg)
    
    def agrupar_met(self):
        return Coleta.met_list(self,self.met)
    
    def rio_arg_list(self, arg_proj):
        b = arg_proj[0]
        for i in range(1, len(arg_proj)):
            b = pd.merge(b, arg_proj[i], on=['gen'], how = 'outer')
        b.columns = self.rios
        return b

    def rio_fec_list(self, fec_proj):
        b = fec_proj[0]
        for i in range(1, len(fec_proj)):
            b = pd.merge(b, fec_proj[i], on=['gen'], how = 'outer')
        b.columns = self.rios
        return b

    def rio_met_list(self, met_proj):
        b = met_proj[0]
        for i in range(1, len(met_proj)):
            #b = pd.DataFrame.join(b, met_proj[i])
            b = pd.merge(b, met_proj[i], left_index=True, right_index=True)
        b.columns = self.rios
        return b
    
class Projeto:

    def __init__(self,nome_projeto, rio_1,rio_2,rio_3,rio_4,rio_5,rio_6,rio_7,rio_8,rio_9,rio_10):
        self.nome_projeto = nome_projeto
        
        self.rios = [rio_1.nome_rio,rio_2.nome_rio,rio_3.nome_rio,rio_4.nome_rio,rio_5.nome_rio,rio_6.nome_rio,
                    rio_7.nome_rio,rio_8.nome_rio,rio_9.nome_rio, rio_10.nome_rio]
        self.arg_proj = [rio_1.agrupar_arg()['mean'],rio_2.agrupar_arg()['mean'],rio_3.agrupar_arg()['mean'],
                        rio_4.agrupar_arg()['mean'],rio_5.agrupar_arg()['mean'],rio_6.agrupar_arg()['mean'],
                        rio_7.agrupar_arg()['mean'],rio_8.agrupar_arg()['mean'],rio_9.agrupar_arg()['mean'],
                        rio_10.agrupar_arg()['mean']]
        
        self.met_proj = [rio_1.agrupar_met()['mean'],rio_2.agrupar_met()['mean'],rio_3.agrupar_met()['mean'],
                        rio_4.agrupar_met()['mean'],rio_5.agrupar_met()['mean'],rio_6.agrupar_met()['mean'],
                        rio_7.agrupar_met()['mean'],rio_8.agrupar_met()['mean'],rio_9.agrupar_met()['mean'],
                        rio_10.agrupar_met()['mean']]
        
        self.fec_proj = [rio_1.agrupar_fec()['mean'],rio_2.agrupar_fec()['mean'],rio_3.agrupar_fec()['mean'],
                        rio_4.agrupar_fec()['mean'],rio_5.agrupar_fec()['mean'],rio_6.agrupar_fec()['mean'],
                        rio_7.agrupar_fec()['mean'],rio_8.agrupar_fec()['mean'],rio_9.agrupar_fec()['mean'],
                        rio_10.agrupar_fec()['mean']]
        
        
        #pode fazer o mesmo para ciano, e fecal
        
    def arg_agrupar_rios(self):
 
        return Rio.rio_arg_list(self,self.arg_proj)
    
    def met_agrupar_rios(self):
 
        return Rio.rio_met_list(self,self.met_proj)

    def fec_agrupar_rios(self):
 
        return Rio.rio_fec_list(self,self.fec_proj)
    
    def arg_graf_rios(self):
        graf_arg = Rio.rio_arg_list(self,self.arg_proj).T
        rios_perc = graf_arg.div(graf_arg.sum(axis="columns"), axis="index")
        rios_perc.plot.barh(stacked=True,color=cores_lista1)
        plt.legend(bbox_to_anchor = (1.05, 1.05))

    def met_graf_rios(self):
        graf_met = Rio.rio_met_list(self,self.met_proj).T
        rios_perc = graf_met.div(graf_met.sum(axis="columns"), axis="index")
        rios_perc.plot.barh(stacked=True,color=cores_lista1)
        plt.legend(bbox_to_anchor = (1.05, 1.05))
        
    def fec_graf_rios(self):
        graf_fec = Rio.rio_fec_list(self,self.fec_proj).T
        rios_perc = graf_fec.div(graf_fec.sum(axis="columns"), axis="index")
        rios_perc.plot.barh(stacked=True,color=cores_lista1)
        plt.legend(bbox_to_anchor = (1.05, 1.05))
```
