```python
#ANOVA
do = big.loc[:,['Data da Coleta','Localidade','Oxi (mg/L)']]
ut = big.loc[:,['Data da Coleta','Localidade','Turbidez']]
hp = big.loc[:,['Data da Coleta','Localidade','ph']]
doc =  big.loc[:,['Data da Coleta','Localidade','COD (mg/L)']]
tn = big.loc[:,['Data da Coleta','Localidade','NT (mg/L)']]
lco = big.loc[:,['Data da Coleta','Localidade','Coliformes Total (NMP/100ml)']]
ce = big.loc[:,['Data da Coleta','Localidade','Escherichia Coli (NMP/100ml)']]
iv = big.loc[:,['Data da Coleta','Localidade','Vibrio (CFU/ml)']]
p_ = big.loc[:,['Data da Coleta','Localidade','P (µg/mL)']]
ef = big.loc[:,['Data da Coleta','Localidade','Fe (µg/mL)']]
sra = big.loc[:,['Data da Coleta','Localidade','As(µg/mL)']]
dc = big.loc[:,['Data da Coleta','Localidade','Cd (µg/mL)']]
oc = big.loc[:,['Data da Coleta','Localidade','Co (µg/mL)']]
rc = big.loc[:,['Data da Coleta','Localidade','Cr (µg/mL)']]
la = big.loc[:,['Data da Coleta','Localidade','Al']]
```

```python
do_pivot = pd.pivot(do, index='Data da Coleta', columns='Localidade', values='Oxi (mg/L)')
ut_pivot = pd.pivot(ut, index='Data da Coleta', columns='Localidade', values='Turbidez')
hp_pivot = pd.pivot(hp, index='Data da Coleta', columns='Localidade', values='ph')
doc_pivot = pd.pivot(doc, index='Data da Coleta', columns='Localidade', values='COD (mg/L)')
tn_pivot = pd.pivot(tn, index='Data da Coleta', columns='Localidade', values='NT (mg/L)')
lco_pivot = pd.pivot(lco, index='Data da Coleta', columns='Localidade', values='Coliformes Total (NMP/100ml)')
ce_pivot = pd.pivot(ce, index='Data da Coleta', columns='Localidade', values='Escherichia Coli (NMP/100ml)')
iv_pivot = pd.pivot(iv, index='Data da Coleta', columns='Localidade', values='Vibrio (CFU/ml)')
p_pivot = pd.pivot(p_, index='Data da Coleta', columns='Localidade', values='P (µg/mL)')
ef_pivot = pd.pivot(ef, index='Data da Coleta', columns='Localidade', values='Fe (µg/mL)')
sra_pivot =pd.pivot(sra, index='Data da Coleta', columns='Localidade', values='As(µg/mL)')
dc_pivot = pd.pivot(dc, index='Data da Coleta', columns='Localidade', values='Cd (µg/mL)')
oc_pivot =pd.pivot(oc, index='Data da Coleta', columns='Localidade', values='Co (µg/mL)')
rc_pivot = pd.pivot(rc, index='Data da Coleta', columns='Localidade', values='Cr (µg/mL)')
la_pivot = pd.pivot(la, index='Data da Coleta', columns='Localidade', values='Al')
```

```python
import scipy.stats as stats
# stats f_oneway functions takes the groups as input and returns ANOVA F and p value
par2 = [do_pivot, ut_pivot, hp_pivot, doc_pivot, tn_pivot, lco_pivot, ce_pivot, iv_pivot, p_pivot,ef_pivot,sra_pivot,dc_pivot,oc_pivot,
       rc_pivot, la_pivot]
anov = []

for i in par2:
    anov.append(stats.f_oneway(i['Centro de Angra'], i['Fazenda Roque'], i['Frade'], i['Jacareí'], 
                               i['Jacuecanga'], i['Mambucaba'],i['Parati'],i['Taquari']))

print(anov)         
```
