```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from scipy import stats
%matplotlib inline
graf = pd.read_excel('Duvidas ft FT.xlsx', sheet_name = 'analises')
```

```python
df_simb = pd.melt(frame=graf,
        id_vars=["Specie","Gutpoint","Reads_total_peixe", "Ano","Rótulos de Linha","soma_patogeno"],
        var_name="Genero", 
        value_name="Reads")
df_simb['%reads'] = df_simb['Reads']/df_simb['Reads_total_peixe']
#sb.barplot(x="%reads", y="Genero", hue="Especie", data=df0, orient="h",errwidth = 0.5)
sb.catplot(x="Gutpoint",y="%reads", hue="Specie",order=['F','MF','M','EM','H'], kind="point", 
            capsize=.15, data=df_simb, aspect=1.5, errwidth=0.5).set(title='Symbionts')
data = {'X': ['F','MF','M','EM','H'], 'Y':[0.00033,0.00044,0.00055,0.00054,0.00053]}
data_1 = {'X1': ['F','MF','M','EM','H'], 'Y1':[0.00021,0.00043,0.0006,0.00059,0.00058]}
cts,cts1 = pd.DataFrame(data=data),pd.DataFrame(data=data_1)
sb.lineplot(data=cts, x='X', y='Y',estimator='max', color='red', legend =False)
sb.lineplot(data=cts1, x='X1', y='Y1',estimator='max', color='green', legend =False)
plt.show()
```

```python
#RANDOM FOREST ACANTHURUS
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
rf = pd.read_excel('PEIXES_ABR.xlsx', sheet_name = 'Planilha5')
y = rf['acanthurus']
X = rf.drop(['sparisoma','kyphosus','scarus','anterior_aa','media_anterior_aa',
            'media_mediana_aa','media_posterior_aa','posterior_aa'],axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 20)
scaler=StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
#feature_list = ['Royal','IDEB','IQ'] 

model = RandomForestRegressor(n_estimators = 30, random_state=30)

model.fit(X_train_scaled, y_train)

y_pred_RF = model.predict(X_test_scaled)

mse_RF = mean_squared_error(y_test, y_pred_RF)
mae_RF = mean_absolute_error(y_test, y_pred_RF)
print('Mean squared error using Random Forest: ', mse_RF)
print('Mean absolute error Using Random Forest: ', mae_RF)
print('r2', model.score(X_test_scaled, y_test))
#Feature ranking...
#import pandas as pd
#feature_list = list(X.columns)
#feature_imp_rf = pd.Series(model.feature_importances_, index=feature_list).sort_values(ascending=False)
```
