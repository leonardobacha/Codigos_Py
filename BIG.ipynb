```python
#ANOVA
do = big.loc[:,['Data da Coleta','Localidade','Oxi (mg/L)']]
ut = big.loc[:,['Data da Coleta','Localidade','Turbidez']]
hp = big.loc[:,['Data da Coleta','Localidade','ph']]
doc =  big.loc[:,['Data da Coleta','Localidade','COD (mg/L)']]
tn = big.loc[:,['Data da Coleta','Localidade','NT (mg/L)']]
lco = big.loc[:,['Data da Coleta','Localidade','Coliformes Total (NMP/100ml)']]
ce = big.loc[:,['Data da Coleta','Localidade','Escherichia Coli (NMP/100ml)']]
iv = big.loc[:,['Data da Coleta','Localidade','Vibrio (CFU/ml)']]
p_ = big.loc[:,['Data da Coleta','Localidade','Fósforo (mg/L)']]
```

```python
# isso aqui é para fazer ANOVA
do_pivot = pd.pivot(do, index='Data da Coleta', columns='Localidade', values='Oxi (mg/L)')
do_pivot.loc['January-22','Gipóia dentro'], do_pivot.loc['January-22','Gipóia fora'] = 5.51 , 6.03
#do_pivot
ut_pivot = pd.pivot(ut, index='Data da Coleta', columns='Localidade', values='Turbidez')
ut_pivot.loc['January-22','Gipóia dentro'], ut_pivot.loc['January-22','Gipóia fora'] = ut_pivot['Gipóia dentro'].mean(),  ut_pivot['Gipóia fora'].mean()

hp_pivot = pd.pivot(hp, index='Data da Coleta', columns='Localidade', values='ph')
hp_pivot.loc['January-22','Gipóia dentro'], hp_pivot.loc['January-22','Gipóia fora'] = hp_pivot['Gipóia dentro'].mean(),  hp_pivot['Gipóia fora'].mean()

doc_pivot = pd.pivot(doc, index='Data da Coleta', columns='Localidade', values='COD (mg/L)')
doc_pivot.loc['January-22','Gipóia dentro'], doc_pivot.loc['January-22','Gipóia fora'] = doc_pivot['Gipóia dentro'].mean(),  doc_pivot['Gipóia fora'].mean()

tn_pivot = pd.pivot(tn, index='Data da Coleta', columns='Localidade', values='NT (mg/L)')
tn_pivot.loc['January-22','Gipóia dentro'], tn_pivot.loc['January-22','Gipóia fora'] = tn_pivot['Gipóia dentro'].mean(),  tn_pivot['Gipóia fora'].mean()

lco_pivot = pd.pivot(lco, index='Data da Coleta', columns='Localidade', values='Coliformes Total (NMP/100ml)')
lco_pivot.loc['January-22','Gipóia dentro'], lco_pivot.loc['January-22','Gipóia fora'] = lco_pivot['Gipóia dentro'].mean(),  lco_pivot['Gipóia fora'].mean()

ce_pivot = pd.pivot(ce, index='Data da Coleta', columns='Localidade', values='Escherichia Coli (NMP/100ml)')
ce_pivot.loc['January-22','Gipóia dentro'], ce_pivot.loc['January-22','Gipóia fora'] = ce_pivot['Gipóia dentro'].mean(),  ce_pivot['Gipóia fora'].mean()


iv_pivot = pd.pivot(iv, index='Data da Coleta', columns='Localidade', values='Vibrio (CFU/ml)')
iv_pivot.loc['January-22','Gipóia dentro'], iv_pivot.loc['January-22','Gipóia fora'] = iv_pivot['Gipóia dentro'].mean(),  iv_pivot['Gipóia fora'].mean()

p_pivot = pd.pivot(p_, index='Data da Coleta', columns='Localidade', values='Fósforo (mg/L)')
p_pivot.loc['January-22','Gipóia dentro'], p_pivot.loc['January-22','Gipóia fora'] = p_pivot['Gipóia dentro'].mean(),  p_pivot['Gipóia fora'].mean()

```

```python
import scipy.stats as stats
# stats f_oneway functions takes the groups as input and returns ANOVA F and p value
par2 = [do_pivot, ut_pivot, hp_pivot, doc_pivot, tn_pivot, lco_pivot, ce_pivot, iv_pivot, p_pivot]
anov = []

for i in par2:
    anov.append(stats.f_oneway(i['Estação Louro'], i['Estação Pablo'], i['Estação Patrícia'], i['Gipóia dentro'], i['Gipóia fora']))

print(anov)             
                #fvalue, pvalue = stats.f_oneway(do_pivot['Estação Louro'], do_pivot['Estação Pablo'], do_pivot['Estação Patrícia'], do_pivot['Gipóia dentro'], do_pivot['Gipóia fora'])
#print(fvalue, pvalue)
# 17.492810457516338 2.639241146210922e-05

# get ANOVA table as R like output
#import statsmodels.api as sm
#from statsmodels.formula.api import ols

# Ordinary Least Squares (OLS) model
#model = ols('value ~ C(treatments)', data=do_pivot_melt).fit()
#anova_table = sm.stats.anova_lm(model, typ=2)
#anova_table
# output (ANOVA F and p value)
  #              sum_sq    df         F    PR(>F)
#C(treatments)  3010.95   3.0  17.49281  0.000026
#Residual        918.00  16.0       NaN       NaN
```
